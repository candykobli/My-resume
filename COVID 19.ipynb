```python
df_country["date"] = pd.to_datetime(df_country["date"])
df_country['month']= df_country["date"].dt.month
df_country['year']= df_country["date"].dt.year
df_country['trimester'] = np.ceil(df_country['month'] / 3)
df_country
```

```python
df_country.set_index('date', inplace=True)
df_country['avg_total_cases']= df_country.groupby(['location', 'trimester'])['new_cases'].transform('mean')
df_country['avg_total_deaths']= df_country.groupby(['location', 'trimester'])['new_deaths'].transform('mean')
df_country['avg_total_vaccinations']= df_country.groupby(['location', 'trimester'])['total_vaccinations'].transform('mean')
df_country['avg_people_vaccinated']= df_country.groupby(['location', 'trimester'])['people_vaccinated'].transform('mean')
df_country['avg_total_cases_base1']= (df_country['avg_total_cases']-df_country['avg_total_cases'].min())/(df_country['avg_total_cases'].max()-df_country['avg_total_cases'].min())
df_country['avg_total_deaths_base1']= (df_country['avg_total_deaths']-df_country['avg_total_deaths'].min())/(df_country['avg_total_deaths'].max()-df_country['avg_total_deaths'].min())
df_country['avg_total_vaccination_base1']= (df_country['avg_total_vaccinations']-df_country['avg_total_vaccinations'].min())/(df_country['avg_total_vaccinations'].max()-df_country['avg_total_vaccinations'].min())
df_country['avg_total_vaccinated_base1']= (df_country['avg_people_vaccinated']-df_country['avg_people_vaccinated'].min())/(df_country['avg_people_vaccinated'].max()-df_country['avg_people_vaccinated'].min())


df_country.reset_index()
```

```python
new_df_country=df_country[['location','year','month','trimester','avg_total_cases','avg_total_deaths','avg_total_vaccinations','avg_people_vaccinated','avg_total_cases_base1','avg_total_deaths_base1','avg_total_vaccination_base1','avg_total_vaccinated_base1']]
new_df_country.reset_index()

```

```python
grouped_df_country = new_df_country.groupby(['location','year','trimester'])['avg_total_cases','avg_total_deaths','avg_total_vaccinations','avg_people_vaccinated','avg_total_cases_base1','avg_total_deaths_base1','avg_total_vaccination_base1','avg_total_vaccinated_base1'].sum()
groupedcountry= grouped_df_country.reset_index()
groupedcountry
```
